```python
lat = requests['latitude']
# Let's say the zips starting with '0' and '1' are okay, for now. (this isn't actually true -- 13221 is in Syracuse, and why?)
is_close = lat.str.startswith('0') | lat.str.startswith('1')
# There are a bunch of NaNs, but we're not interested in them right now, so we'll say they're False
is_far = ~(is_close) & lat.notnull()
```

```python
na_values = ['NO CLUE', 'N/A','NaN', '0']
requests = pd.read_csv('exercise-5/complete.csv', 
                       na_values=na_values, 
                       dtype={'latitude': str},error_bad_lines=False)
```

```python
def fix_zip_codes(lat):

    # Set 00000 zip codes to nan
    zero_lat = lat == '0'
    lat[zero_lat] = np.nan
    
    return lat
```

```python
requests['latitude'] = fix_zip_codes(requests['latitude'])
```

```python
requests['latitude'].unique()
```

```python
na_values = ['NO CLUE', 'N/A','NaN', '0']
request = pd.read_csv('exercise-5/complete.csv', 
                       na_values=na_values, 
                       dtype={'longitude': str},error_bad_lines=False)
```

```python
def fix_long_codes(long):

    # Set 00000 zip codes to nan
    zero_long = long == '0'
    long[zero_long] = np.nan
    
    return long
```

```python
request['longitude'] = fix_long_codes(request['longitude'])
```
